Create the following Series and do the specified
operations:
a) EngAlph, having 26 elements with the alphabets
as values and default index values.
b) Vowels, having 5 elements with index labels ‘a’,
‘e’, ‘i’, ‘o’ and ‘u’ and all the five values set to zero.
Check if it is an empty series.
c) Friends, from a dictionary having roll numbers of
five of your friends as data and their first name
as keys.
d) MTseries, an empty Series. Check if it is an empty
series.
e) MonthDays, from a numpy array having the
number of days in the 12 months of a year. The
labels should be the month numbers from 1 to 12.

```python
import pandas as pd
import numpy as np
import string

# a) EngAlph: 26 alphabets with default index
EngAlph = pd.Series(list(string.ascii_lowercase))
print("a) EngAlph Series:\n", EngAlph, "\n")

# b) Vowels: index labels 'a', 'e', 'i', 'o', 'u' with all values 0
Vowels = pd.Series(0, index=['a', 'e', 'i', 'o', 'u'])
print("b) Vowels Series:\n", Vowels)
print("Is Vowels Series empty?", Vowels.empty, "\n")

# c) Friends: Dictionary of 5 friends
friends_dict = {
    "Alice": 101,
    "Bob": 102,
    "Charlie": 103,
    "David": 104,
    "Eve": 105
}
Friends = pd.Series(friends_dict)
print("c) Friends Series:\n", Friends, "\n")

# d) MTseries: Empty Series and check if it's empty
MTseries = pd.Series(dtype=float)
print("d) MTseries:\n", MTseries)
print("Is MTseries empty?", MTseries.empty, "\n")

# e) MonthDays: from numpy array, labels as 1 to 12
days_array = np.array([31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31])
MonthDays = pd.Series(days_array, index=range(1, 13))
print("e) MonthDays Series:\n", MonthDays)
```

11.Collect data about colleges in Delhi University or any
other university of your choice and number of courses
they run for Science, Commerce and Humanities,
store it in a CSV file and present it using a bar plot.

```python
import csv

# Sample course data for CBIT departments (mocked for this exercise)
data = [
    ["CBIT", "Science", 10],
    ["CBIT", "Commerce", 3],
    ["CBIT", "Humanities", 5]
]

# Write to CSV
with open("cbit_courses.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["College", "Stream", "Courses"])
    writer.writerows(data)

print("CSV file 'cbit_courses.csv' created successfully.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV
df = pd.read_csv("cbit_courses.csv")

# Plot bar graph
plt.figure(figsize=(7, 4))
plt.bar(df["Stream"], df["Courses"], color='skyblue')

# Add labels and title
plt.title("Number of Courses Offered by CBIT")
plt.xlabel("Stream")
plt.ylabel("Number of Courses")
plt.tight_layout()
plt.show()
```

14.Collect the minimum and maximum temperature
of your city for a month and present it using a
histogram plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the temperature CSV file
df = pd.read_csv("city_temperature.csv")

# Plot histograms for Min and Max Temperature
plt.figure(figsize=(10, 5))
plt.hist(df["Min Temperature"], bins=6, alpha=0.7, label='Min Temperature', color='skyblue')
plt.hist(df["Max Temperature"], bins=6, alpha=0.7, label='Max Temperature', color='salmon')

plt.title("City Temperature Histogram for the Month")
plt.xlabel("Temperature (°C)")
plt.ylabel("Number of Days")
plt.legend()
plt.tight_layout()
plt.show()
```

##Basics

```python
#Importing datasets
from sklearn import datasets
iris = datasets.load_iris()
print(iris.data.shape)
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

#Load dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

#Split dataset into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
#a random order instead of selecting from a sorted order, for proper training of the model
#random_state makes it so that the same set of records are used for training while training many times

#Create and train the model
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)

#Make predictions
y_pred = clf.predict(X_test)

#Evaluate the model
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy}")
```
